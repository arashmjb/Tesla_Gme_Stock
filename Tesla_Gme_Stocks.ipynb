```python
import yfinance as yf

# Extract stock data for Tesla (TSLA)
tesla_data = yf.download('TSLA', start='2024-01-01', end='2024-06-01')

# Reset index
tesla_data.reset_index(inplace=True)

# Save the DataFrame to a file (optional)
tesla_data.to_csv('tesla_stock_data.csv', index=False)

# Display the first five rows
print(tesla_data.head())
```

```python
import pandas as pd

# Create DataFrame
tesla_revenue_df = pd.DataFrame(tesla_data)

# Display the last five rows
print(tesla_revenue_df.tail())
```

```python
# Extract stock data for GameStop (GME)
gme_data = yf.download('GME', start='2024-01-01', end='2024-06-01')

# Reset index
gme_data.reset_index(inplace=True)

gme_data.to_csv('gme_stock_data.csv', index=False)

# Display the first five rows
print(gme_data.head())
```

```python
# Display the last five rows of the gme_revenue dataframe
print(gme_data.tail())
```

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Closing Price ($)')
    plt.grid(True)
    plt.show()

# Plot Tesla stock graph
make_graph(tesla_data, title='Tesla Stock Price (2024)')
```

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], color='red')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Closing Price ($)')
    plt.grid(True)
    plt.show()

# Plot GameStop stock graph
make_graph(gme_data, title='GameStop Stock Price (2024)')
```
